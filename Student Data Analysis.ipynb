```python
import pandas as pd
import numpy as np
```

### Data Preparation

```python
# Load the student marks and attendance data from the provided Excel files into two separate Pandas DataFrames.
df1= pd.read_excel("C:\\Users\\HP\\OneDrive\\Documents\\student performance analysis.xlsx", sheet_name="Marks")
```

```python
df2= pd.read_excel("C:\\Users\\HP\\OneDrive\\Documents\\student performance analysis.xlsx", sheet_name="Attendance")
```

```python
df2.head(5)
```

```python
df2.tail(5)
```

```python
# Merge the two DataFrames on the Name column to create a single DataFrame.
df=pd.merge(df1,df2, on='Name')
```

```python
# Replace attendance values ('Y' or 'N') with numeric values (1 for 'Y' and 0 for 'N').
attendance_columns = [col for col in df.columns if "Attendance" in col]
df[attendance_columns] = df[attendance_columns].replace({'Y':1,'N':0})
```

```python
df.head(5)
```

```python
#Deal with the missing values
import numpy as np

df.replace(to_replace='^\s*$', value=np.nan, regex=True,inplace=True)
```

```python
#Deal with the missing values
df.isnull().sum()
```

```python
df.fillna(0,inplace=True)
```

```python
df.isnull().sum()
```

```python
df.head(30)
```

```python
#	Do the column cleaning like name column is the combination of upper and lower case.
df['Name'] = df['Name'].apply(lambda x: x.lower())
df.head(10)
```

### Data Transformation

```python
# Total marks for each student.
marks_columns=["Mini Test 1","Mini Test 2", "Live Test","Assignment"]
df["Total_Marks"]=df[marks_columns].astype(float).sum(axis=1)
```

```python
df.head(7)
```

```python
marks_columns=["Mini Test 1","Mini Test 2", "Live Test","Assignment"]
total =df[marks_columns].max().sum()
df["Percentage_Marks"]=((df["Total_Marks"]/total)*100).round(2)
```

```python
df.head(5)
```

```python
# 	Attendance percentage for each student based on attendance columns.
Attendence_All=["Attendance Day 1","Attendance Day 2","Attendance Day 3","Attendance Day 4","Attendance Day 5"]
df["Total_attendence"]=df[Attendence_All].astype(float).sum(axis=1)
totall =df[Attendence_All].max().sum()
df["Percentage_attendence"]=((df["Total_attendence"]/totall)*100).round(2)
```

```python
df.head(5)
```

```python
# Calculate the weighted percentage  Attendance(40%), Mini test01(10%), Mini test02(10%), Live test(20%), Assignment (20%).
columns= ['Mini Test 1', 'Mini Test 2', 'Live Test', 'Assignment']
df[columns] = df[columns].apply(pd.to_numeric, errors='coerce')

df['Weighted Percentage'] = (
    (df['Percentage_attendence'] * 0.40) +
    (df['Mini Test 1'] * 0.10) +
    (df['Mini Test 2'] * 0.10) +
    (df['Live Test'] * 0.20) +
    (df['Assignment'] * 0.20)
)
df['Weighted Percentage'] = df['Weighted Percentage'].round(2)
```

```python
df.head(5)
```

```python
#	Classify each student's performance into categories
conditions = [
    (df['Weighted Percentage'] >= 85),
    (df['Weighted Percentage'] >= 71) & (df['Weighted Percentage'] <= 84),
    (df['Weighted Percentage'] >= 50) & (df['Weighted Percentage'] <= 70),
    (df['Weighted Percentage'] < 50)
]
performance = ['Excellent', 'Good', 'Average', 'Needs Improvement']
df['Performance Category'] = np.select(conditions, performance)
```

```python
df.head(5)
```

### Visualization

```python
# Create a bar chart displaying weighted percentages for top 5 students.
import matplotlib.pyplot as plt
top_5= df.sort_values(by='Weighted Percentage' , ascending=False).head(5)
top_5.plot(x='Name',y='Weighted Percentage',kind='bar',legend=False,color='#DE3163',title='Top 5 Students by Weighted Percentage')
plt.ylabel('Weighted Percentage')
plt.xlabel('Student Name')
plt.show()
```

```python
# Create a pie chart showing the distribution of students across the four performance categories.
category = df['Performance Category'].value_counts()
category.plot(kind='pie',autopct='%1.1f%%', startangle=90, colors=['#FF7F50', '#DFFF00', '#FFBF00','#DE3163' ], title='Student Performance Distribution',ylabel='')
plt.tight_layout()
plt.show()
```

```python
# Create a chart to show the students where attendance is less than 50
low_attendance = df[df["Percentage_attendence"] < 50]
plt.figure(figsize=(10, 10))
plt.bar(low_attendance["Name"], low_attendance["Percentage_attendence"], color="blue")
plt.title("Students with Attendance Less Than 50%")
plt.xlabel("Student Name")
plt.ylabel("Attendance (%)")
plt.xticks(rotation=90)
plt.show()
```

```python
#	Create box plots for each test (Live Test, Mini Test 1, Mini Test 2, Assignment) to visualize the spread and detect potential outliers in scores.
test_columns = ['Live Test', 'Mini Test 1', 'Mini Test 2', 'Assignment']
df[test_columns].plot(kind='box',figsize=(10, 6),patch_artist=True,boxprops=dict(facecolor='#40E0D0', color='black'),medianprops=dict(color='red'),flierprops=dict(marker='o', markerfacecolor='orange', markersize=6, linestyle='none'),title='Box Plots of Test Scores')
plt.ylabel('Scores')
plt.tight_layout()
plt.show()
```

```python
# Create attendance bins
df['Attendance Group'] = pd.cut(
    df['Percentage_attendence'], 
    bins=[0, 50, 60, 75, 90, 100], 
    labels=['<50%', '50-60%', '60-75%', '75-90%', '90-100%']
)

# Group and calculate average weighted percentage
attendance_group_perf = df.groupby('Attendance Group')['Weighted Percentage'].mean().sort_index()

# Plot
attendance_group_perf.plot(kind='bar', color='#26a69a', title='Avg Weighted % by Attendance Group')
plt.ylabel('Average Weighted Percentage')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
